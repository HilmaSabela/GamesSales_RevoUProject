```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as date
```

```python
#Read dataset
df = pd.read_excel(r"G:\Data Analyst\Project_11_GamesSales\Games Sales - Case Study.xlsx")
df
```

# Data Cleaning

```python
# Delete duplicates data
df.drop_duplicates()
```

```python
#Replace Nan Value with empty string
df['Series'] = df['Series'].fillna('')
df
```

# Data Plot

```python
# Which game is the newest and the oldest game in the dataset?

df['Release'].min()
```

```python
df['Release'].max()
```

```python
# Which publisher published most of games

publisher = df['Publisher'].value_counts().sort_values(ascending=True).plot(kind = 'barh', figsize = (10,18),title = 'Publishers Who Publish the Most')

publisher.set_ylabel("Publisher")
publisher.set_xlabel("Total")
```

```python
#Which developer developed most games
publisher = df['Developer'].value_counts().sort_values(ascending=True).plot(kind = 'barh', figsize = (10,18),title = 'Developer Who Developed the Most')

publisher.set_ylabel("Developer")
publisher.set_xlabel("Total")
```

```python
#Which Series has the most sales

df.pivot_table(values = 'Sales', index = 'Series', aggfunc='sum').sort_values(by = 'Sales', ascending=True).plot(kind='barh', figsize = (10,18), title = 'The Most Sales of Series')
plt.tight_layout()
plt.show()
```

```python
#Which Series has the most Games

df.pivot_table(values = 'Name', index = 'Series', aggfunc='count').sort_values(by = 'Name', ascending=True).plot(kind='barh', figsize = (10,18), title ='Series with the Games')

plt.tight_layout()
plt.show()

```
